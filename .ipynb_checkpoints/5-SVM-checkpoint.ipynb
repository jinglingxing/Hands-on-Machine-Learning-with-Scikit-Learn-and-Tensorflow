### Exercise 8
Train a LinearSVC on a linearly separable dataset. Then train an SVC and a SGDClassifier on the same dataset. See if you can get them to produce the same model

```python
import numpy as np
from sklearn import datasets
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.svm import LinearSVC
from sklearn.linear_model import SGDClassifier
```

```python
from sklearn.svm import SVC
from sklearn import datasets

iris = datasets.load_iris()
X = iris["data"][:, (2, 3)]  # petal length, petal width
y = iris["target"]

setosa_or_versicolor = (y == 0) | (y == 1)
X = X[setosa_or_versicolor]
y = y[setosa_or_versicolor]
```

```python
lin_clf = Pipeline([
        ("scaler", StandardScaler()),
        ("linear_svc", LinearSVC(C=1, loss="hinge", random_state=42)),
    ])

lin_clf.fit(X, y)
```

```python
sgd_clf = Pipeline([
    ("scaler", StandardScaler()),
    ("sgd_classifier", SGDClassifier(loss="hinge"))
])
```

```python
sgd_clf.fit(X, y)
```

```python
svm_clf = Pipeline([
        ("scaler", StandardScaler()),
        ("linear_svc", SVC(kernel="linear", C=1))
    ])

svm_clf.fit(X, y)
```

```python
lin_clf['linear_svc']
```

```python
print("LinearSVC:                   ", lin_clf['linear_svc'].intercept_, lin_clf['linear_svc'].coef_)
print("SVC:                         ", svm_clf['linear_svc'].intercept_, svm_clf['linear_svc'].coef_)
print("SGDClassifier(alpha={:.5f}):".format(sgd_clf['sgd_classifier'].alpha), sgd_clf['sgd_classifier'].intercept_, sgd_clf['sgd_classifier'].coef_)
```

```python
import matplotlib.pyplot as plt
# Compute the slope and bias of each decision boundary
w1 = -lin_clf['linear_svc'].coef_[0, 0]/lin_clf['linear_svc'].coef_[0, 1]
b1 = -lin_clf['linear_svc'].intercept_[0]/lin_clf['linear_svc'].coef_[0, 1]
w2 = -svm_clf['linear_svc'].coef_[0, 0]/svm_clf['linear_svc'].coef_[0, 1]
b2 = -svm_clf['linear_svc'].intercept_[0]/svm_clf['linear_svc'].coef_[0, 1]
w3 = -sgd_clf['sgd_classifier'].coef_[0, 0]/sgd_clf['sgd_classifier'].coef_[0, 1]
b3 = -sgd_clf['sgd_classifier'].intercept_[0]/sgd_clf['sgd_classifier'].coef_[0, 1]

# Transform the decision boundary lines back to the original scale
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
line1 = scaler.inverse_transform([[-10, -10 * w1 + b1], [10, 10 * w1 + b1]])
line2 = scaler.inverse_transform([[-10, -10 * w2 + b2], [10, 10 * w2 + b2]])
line3 = scaler.inverse_transform([[-10, -10 * w3 + b3], [10, 10 * w3 + b3]])

# Plot all three decision boundaries
plt.figure(figsize=(11, 4))
plt.plot(line1[:, 0], line1[:, 1], "k:", label="LinearSVC")
plt.plot(line2[:, 0], line2[:, 1], "b--", linewidth=2, label="SVC")
plt.plot(line3[:, 0], line3[:, 1], "r-", label="SGDClassifier")
plt.plot(X[:, 0][y==1], X[:, 1][y==1], "bs") # label="Iris versicolor"
plt.plot(X[:, 0][y==0], X[:, 1][y==0], "yo") # label="Iris setosa"
plt.xlabel("Petal length", fontsize=14)
plt.ylabel("Petal width", fontsize=14)
plt.legend(loc="upper center", fontsize=14)
plt.axis([0, 5.5, 0, 2])

plt.show()
```

### Exercise 9
Train an SVM classifier on the MNIST dataset. Since SVM classifiers are binary classifiers, you
will need to use one-versus-all to classify all 10 digits. You may want to tune the hyperparameters
using small validation sets to speed up the process. What accuracy can you reach?

```python
from sklearn.datasets import fetch_openml
mnist = fetch_openml('mnist_784', version=1, cache=True, as_frame=False)

X = mnist["data"]
y = mnist["target"].astype(np.uint8)
```

```python
X_train = X[:60000]
y_train = y[:60000]
X_test = X[60000:]
y_test = y[60000:]
```

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train.astype(np.float32))
X_test_scaled = scaler.transform(X_test.astype(np.float32))
```

```python
from sklearn.model_selection import RandomizedSearchCV
from scipy.stats import reciprocal, uniform
svm_clf = SVC(gamma="scale")
param_distributions = {"gamma": reciprocal(0.001, 0.1), "C": uniform(1, 10)}
rnd_search_cv = RandomizedSearchCV(svm_clf, param_distributions, n_iter=10, verbose=2, cv=3)
rnd_search_cv.fit(X_train_scaled[:1000], y_train[:1000])
```
