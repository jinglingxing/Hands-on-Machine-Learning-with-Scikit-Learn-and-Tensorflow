```python
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.model_selection import StratifiedShuffleSplit
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt
from sklearn.metrics import silhouette_samples
from matplotlib.ticker import FixedLocator, FixedFormatter
import matplotlib as mpl
import numpy as np
from sklearn.decomposition import PCA
```

```python
# import data
olivetti = datasets.fetch_olivetti_faces()
```

```python
### Question 10
# stratified shuffle
split = StratifiedShuffleSplit(n_splits=1, test_size=40, random_state=42)
for train_valid_index, test_index in split.split(olivetti.data, olivetti.target):
    X_train_valid, X_test = olivetti.data[train_valid_index], olivetti.data[test_index]
    y_train_valid, y_test = olivetti.target[train_valid_index], olivetti.target[test_index]
    
split = StratifiedShuffleSplit(n_splits=1, test_size=80, random_state=42)
for train_index, valid_index in split.split(X_train_valid, y_train_valid):
    X_train, X_valid = X_train_valid[train_index], X_train_valid[valid_index]
    y_train, y_valid = y_train_valid[train_index], y_train_valid[valid_index]
    
print(X_train.shape, y_train.shape)
print(X_valid.shape, y_valid.shape)
print(X_test.shape, y_test.shape)
```

```python
# reduce the demensionality
pca = PCA(0.99)
X_train_pca = pca.fit_transform(X_train)
X_valid_pca = pca.transform(X_valid)
X_test_pca = pca.transform(X_test)
pca.n_components_
```

```python
# find the optimal k using silhoutette score
k_range = range(5, 200, 5)
kmeans_per_k = [KMeans(n_clusters=k).fit(X_train_pca) for k in k_range]    
silhouette_scores = [silhouette_score(X_train_pca, model.labels_) for model in kmeans_per_k]

plt.figure(figsize=(8, 3))
best_index = np.argmax(silhouette_scores)
best_k = k_range[best_index]
best_score = silhouette_scores[best_index]
plt.plot(k_range, silhouette_scores, "bo-")
plt.xlabel("$k$", fontsize=14)
plt.ylabel("Silhouette score", fontsize=14)
plt.plot(best_k, best_score, "rs")
plt.show()
```

```python
# we expect 40 clusters but the best k is 120. A person might have different faces based on different face expressions
inertias = [model.inertia_ for model in kmeans_per_k]
best_inertias = inertias[best_index]
plt.plot(k_range, inertias, "bo-")
plt.xlabel("$k$", fontsize=14)
plt.ylabel("Inertias", fontsize=14)
plt.plot(best_k, best_inertias, "rs")
plt.show()
```

```python
# visualize the clusters
best_model = kmeans_per_k[best_index]
best_model
```

```python
def plot_faces(faces, labels, n_cols=5):
    faces = faces.reshape(-1, 64, 64)
    n_rows = (len(faces) - 1) // n_cols + 1
    plt.figure(figsize=(n_cols, n_rows * 1.1))
    for index, (face, label) in enumerate(zip(faces, labels)):
        plt.subplot(n_rows, n_cols, index + 1)
        plt.imshow(face, cmap="gray")
        plt.axis("off")
        plt.title(label)
    plt.show()

for cluster_id in np.unique(best_model.labels_):
    print("Cluster", cluster_id)
    in_cluster = best_model.labels_==cluster_id
    faces = X_train[in_cluster]
    labels = y_train[in_cluster]
    plot_faces(faces, labels)
```

```python
from sklearn.ensemble import RandomForestClassifier
# Train a classifier to predict which person is represented in each picture
# Evaluate it on the validation set
clf = RandomForestClassifier(n_estimators=150, random_state=42)
clf.fit(X_train_pca, y_train)
clf.score(X_valid_pca, y_valid)
```

```python
# Use KMeans as a dimentionality reduction tool
X_train_reduced = best_model.transform(X_train_pca)
X_valid_reduced = best_model.transform(X_valid_pca)
X_test_reduced = best_model.transform(X_test_pca)
```

```python
clf = RandomForestClassifier(n_estimators=150, random_state=42)
clf.fit(X_train_reduced, y_train)
clf.score(X_valid_reduced, y_valid)
```

```python
# Search for the number of clusters that allows the classifier to get the best performance
from sklearn.model_selection import GridSearchCV
param_grid = [
    {'n_estimators': [3, 10, 30, 60], 'max_features': [2,4,6,8,10]},
    {'bootstrap': [False], 'n_estimators': [3, 10], 'max_features': [2, 3, 4]},
]
```

```python
forest_clf = RandomForestClassifier()
grid_search = GridSearchCV(forest_clf, param_grid, cv=5,
                           scoring='neg_mean_squared_error',
                           return_train_score=True)
```

```python
grid_search.fit(X_train_reduced, y_train)
```

```python
grid_search.best_params_
```

We could use a GridSearchCV like we did earlier in this notebook, but since we already have a validation set, we don't need K-fold cross-validation, and we're only exploring a single hyperparameter, so it's simpler to just run a loop manually:

```python
from sklearn.pipeline import Pipeline
for k in k_range:
    pipeline = Pipeline([
        ('kmeans', KMeans(n_clusters=k, random_state=42)),
        ('random_forest', RandomForestClassifier(n_estimators=150, random_state=42))
    ])
    pipeline.fit(X_train_pca, y_train)
    print(k, pipeline.score(X_valid_pca, y_valid))
```

Oh well, even by tuning the number of clusters, we never get beyond 80% accuracy. Looks like the distances to the cluster centroids are not as informative as the original images

```python
# What if you append the features from the reduced set to the origin
X_train_extended = np.c_[X_train_pca, X_train_reduced]
X_valid_extended = np.c_[X_valid_pca, X_valid_reduced]
X_test_extended = np.c_[X_test_pca, X_test_reduced]
```

```python
### Question 12
# Train a Gaussian Mixture Model
# Reduce the dataset's dimensionality(e.g., use PCA, preserving 99% of variance)
from sklearn.mixture import GaussianMixture
gm = GaussianMixture(n_components=40, n_init=10)
gm.fit(X_train_pca)
```

```python
gm.weights_
```

```python
# Generate some new faces and visualize them
X_train_new, y_train_new = gm.sample(n_samples=20)
X_train_new_origin = pca.inverse_transform(X_train_new)
plot_faces(X_train_new_origin, y_train_new)
```

```python
# Try to modify some images(e.g.,rotate, flip, darken) and see if the model can detect the anomalies
n_rotated = 4
rotated = np.transpose(X_train[:n_rotated].reshape(-1, 64, 64), axes=[0, 2, 1])
rotated = rotated.reshape(-1, 64*64)
y_rotated = y_train[:n_rotated]

n_flipped = 3
flipped = X_train[:n_flipped].reshape(-1, 64, 64)[:, ::-1]
flipped = flipped.reshape(-1, 64*64)
y_flipped = y_train[:n_flipped]

n_darkened = 3
darkened = X_train[:n_darkened].copy()
darkened[:, 1:-1] *= 0.3
y_darkened = y_train[:n_darkened]

X_bad_faces = np.r_[rotated, flipped, darkened]
y_bad = np.concatenate([y_rotated, y_flipped, y_darkened])

plot_faces(X_bad_faces, y_bad)
```

```python
X_bad_faces_pca = pca.transform(X_bad_faces)
```

```python
gm.predict(X_bad_faces_pca)
```

```python
gm.score_samples(X_bad_faces_pca)
```

```python
gm.score_samples(X_train_pca[:10])
```

The bad faces are all considered highly unlikely by the Gaussian Mixture model. 

```python
# Question 13
# Compute the reconstruction error for each image
def reconstruct_error(pca, X):
    X_pca = pca.transform(X)
    X_reconstructed = pca.inverse_transform(X_pca)
    mse = np.square(X_reconstructed - X).mean()
    return mse
```

```python
reconstruct_error(pca, X_train).mean()
```

```python
reconstruct_error(pca, X_bad_faces)
```
